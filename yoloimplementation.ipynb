```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D, Input, BatchNormalization, ZeroPadding2D, Activation
from tensorflow.keras.models import Model, Sequential
```

```python
archConfig = [
    (64,7,2,3), #(numFilters, kernelSize, stride, padding)
    "MaxPooling", #maxpooling(2x2,stride = 2)
    (192,3,1,1),
    "MaxPooling",
    (128,1,1,0),
    (256,3,1,1),
    (256,1,1,0),
    (512,3,1,1),
    "MaxPooling",
    (256,1,1,0),
    (512,3,1,1),
    (256,1,1,0),
    (512,3,1,1),
    (256,1,1,0),
    (512,3,1,1),
    (256,1,1,0),
    (512,3,1,1),
    (512,1,1,0),
    (1024,3,1,1),
    "MaxPooling",
    (512,1,1,0),
    (1024,3,1,1),
    (512,1,1,0),
    (1024,3,1,1),
    (1024,3,1,1),
    (1024,3,2,1),
    (1024,3,1,1),
    (1024,3,1,1),   
]
```

```python
def CNNblock(outChannels, kernalSize = 3, strides = 1, pad = 0):
    CNNblock = Sequential([
        ZeroPadding2D(padding =(pad,pad)),
        Conv2D(outChannels, kernel_size = kernalSize, strides = strides, use_bias = False),
        BatchNormalization(),
        Activation("relu")]
    )
    return CNNblock
```

```python
def YOLOv1(archConfig):
    model = Sequential()
    for x in archConfig:
        if x == "MaxPooling":
            model.add(MaxPooling2D(pool_size = (2,2), strides = 2))
        else:
            model.add(CNNblock(x[0], x[1], x[2], x[3]))
    model.add(Flatten())
    model.add(Dense(4096, activation = "relu"))
    model.add(Dropout(0.5))
    model.add(Dense(496, activation = "relu"))
    
    model.compile(optimizer = "adam")
    return model
            
```

```python
model = YOLOv1(archConfig)
model.build((None, 32,32,3))
model.summary()
```
